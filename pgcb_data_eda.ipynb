```python
import re 

def sorted_nicely( l ): 
    #https://stackoverflow.com/questions/2669059/how-to-sort-alpha-numeric-set-in-python
    """ Sort the given iterable in the way that humans expect.""" 
    convert = lambda text: int(text) if text.isdigit() else text 
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(l, key = alphanum_key)
```

```python
df = pd.DataFrame()
i = 0
yy_path = os.path.join(dataset_directory, yy_folder[i]) 
os.chdir(yy_path)
months = os.listdir()
months = sorted_nicely(months)
for mm in months:
    mm_path = os.path.join(yy_path, mm)
    os.chdir(mm_path)
    days = os.listdir()
    days = sorted_nicely(days)
    for dd in days:
        dd_path = os.path.join(mm_path, dd)
        dd = dd.split(".")[0]
        temp_df = pd.read_csv(dd_path)
        required_column = temp_df.iloc[:, 4][2 : 50]
        #print(required_column)
        df[dd] = required_column
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
from tslearn.clustering import silhouette_score

plt.rcParams['figure.figsize'] = [25, 8]

from tslearn.clustering import TimeSeriesKMeans
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
```

```python
seed = 0
np.random.seed(seed)
X_train = TimeSeriesScalerMeanVariance().fit_transform(df_t)
sz = X_train.shape[1]
```

```python
X_train.shape
```

```python
Sum_of_squared_distances = []
K = range(2,12)
for k in K:
    km = TimeSeriesKMeans(n_clusters=k,
                          n_init=2,
                          metric="dtw",
                          verbose=False,
                          max_iter_barycenter=10,
                          random_state=0)
    
    km = km.fit(X_train)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method for Optimal k Using Dynamic Time Warping')
plt.show()
```

```python
Sum_of_squared_distances = []
K = range(2,12)
for k in K:
    km = TimeSeriesKMeans(n_clusters=k,
                          n_init=2,
                          metric="euclidean",
                          verbose=False,
                          max_iter_barycenter=10,
                          random_state=0)
    
    km = km.fit(X_train)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method for Optimal k Using Euclidean Distance')
plt.show()
```

```python
# Euclidean k-means

n_clusters = 7 # 4 and 7 give best silhoutte scores, 0.33

print("Euclidean k-means")
km = TimeSeriesKMeans(n_clusters=n_clusters, verbose=False, random_state=seed)
y_pred_km = km.fit_predict(X_train)


print("Euclidean silhoutte: {:.2f}".format(silhouette_score(X_train, y_pred_km, metric="euclidean")))

plt.figure()
for yi in range(n_clusters):
    plt.subplot(3, n_clusters, yi + 1)
    for xx in X_train[y_pred_km == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)
    plt.plot(km.cluster_centers_[yi].ravel(), "r-")
    plt.xlim(0, sz)
    plt.ylim(-4, 4)
    plt.text(0.55, 0.85,'Cluster %d' % (yi + 1),
             transform=plt.gca().transAxes)
    if yi == 1:
        plt.title("Euclidean $k$-means")
```

```python
# DBA-k-means

n_clusters = 6

print("DBA k-means")
dba_km = TimeSeriesKMeans(n_clusters=n_clusters,
                          n_init=2,
                          metric="dtw",
                          verbose=False,
                          max_iter_barycenter=10,
                          random_state=seed)
y_pred_dba_km = dba_km.fit_predict(X_train)
print("DBA silhoutte: {:.2f}".format(silhouette_score(X_train, y_pred_dba_km, metric="dtw")))

for yi in range(n_clusters):
    plt.subplot(3, n_clusters, yi+1)
    
    for xx in X_train[y_pred_dba_km == yi]:
        plt.plot(xx.ravel(), "k-", alpha=.2)

    plt.plot(dba_km.cluster_centers_[yi].ravel(), "r-")
    plt.xlim(0, sz)
    plt.ylim(-4, 4)
    plt.text(0.55, 0.85,'Cluster %d' % (yi + 1), transform=plt.gca().transAxes)
    
    if yi == 1:
        plt.title("DBA $k$-means")
        
plt.tight_layout()
plt.show()
```

```python
quantization_level = [float(np.amin(avg_temp.iloc[:, [0]])), 24, 28, 31, 32, 34, float(np.amax(avg_temp.iloc[:, [0]]))]

'''
for i in range(seasons):
    new_level = np.array(avg_temp.iloc[[i], [0]])[0][0]
    quantization_level.append(new_level)
'''
    
quantization_level = np.array(quantization_level)
print(quantization_level)

avg_temp["label"] = 0
for i in range(avg_temp.shape[0]):
    for k, j in enumerate(quantization_level):
        if k < (len(quantization_level) - 1):
            delta = abs(quantization_level[k] - quantization_level[k+1])
        elif k == (len(quantization_level) - 1):
            delta = 0
            
        print(k, delta)
        
        mul = 0.05
        if (np.array(avg_temp.iloc[[i], [0]])[0][0] >= j) and (np.array(avg_temp.iloc[[i], [0]])[0][0] <= (j + mul*(delta/2))):
            avg_temp.iloc[[i], [1]] = k
            break
        elif (np.array(avg_temp.iloc[[i], [0]])[0][0] > (j + mul*(delta/2))) and (np.array(avg_temp.iloc[[i], [0]])[0][0] <= (j + (1 - mul)*delta)):
            avg_temp.iloc[[i], [1]] = k + 1
            break
```
